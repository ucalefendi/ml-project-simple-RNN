```python
from tensorflow.keras.preprocessing.text import  one_hot
```

```python
#sentences

sentence = [
    'The cat sits on the mat',
    'She loves playing guitar',
    'He enjoys reading books',
    'The sun is shining bright',
    'Birds are flying in the sky',
    'I like to drink coffee',
    'They went to the park',
    'The dog barked loudly'
]
```

```python
#define vocabulary size

voc_size = 10000
```

```python
#One Hot representation for every word
one_hot_repr = [one_hot(words,voc_size) for words in sentence]
```

```python
#word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np

```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding="post",maxlen=sent_length)
print(embedded_docs)
```

```python
#feature representation
dim=10 #dim=dimension
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))

model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
